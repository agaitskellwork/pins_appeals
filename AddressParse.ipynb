```python
address_df = pd.DataFrame(data=parsed)
address_df.columns=['index','value','column']
address_df
```

```python
pivot_table = address_df.pivot_table(index=['index'],
                                     columns='column', 
                                     values='value',
                                     aggfunc=lambda x: ' '.join(x))
pivot_table = pivot_table.reset_index()
pivot_table.shape
```

```python
dataset = pd.merge(pivot_table,appeals_df,on='index')
dataset.shape
```

```python
dataset['postcode'] = dataset['postcode'].str.upper()
dataset['postcode']
```
